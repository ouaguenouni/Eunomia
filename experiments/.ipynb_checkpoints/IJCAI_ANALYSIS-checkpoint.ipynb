```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from Eunomia.experiments import *
```

```python
dicts = load_yaml_files_from_directory("SX_IJCAI_M/2024-01-10")
print(f"Remaining {len(dicts)}")
```

```python
df = {
    "time":[],
    "n":[],
    "m":[],
    "ratio":[],
    "w_rhat":[],
}
for d in dicts:
    #print("==========")
    #print(d.keys())
    #print(f"n = {d['n']}, m = {d['m']}, ratio = {np.log2(d['m'])/d['n']}")

    df["time"].append(d["time"])
    df["n"].append(d["n"])
    df["m"].append(d["m"])
    df["ratio"].append(np.log2(d['m'])/d['n'])
    df["w_rhat"].append(np.array(d["w_rhat"]).mean())
    
df = pd.DataFrame(df)
df.m = df.m // 3
df.m = df.m * 3
df["w_rhat"].mean()
```

```python
sns.lineplot(x = "m", y = "time", data = df)
plt.savefig("M_TIME.png")
```

```python
dicts = load_yaml_files_from_directory("SX_IJCAI_HYPERPARAMETER_3/2024-01-10")
dicts = dicts + load_yaml_files_from_directory("SX_IJCAI_HYPERPARAMETER_3/2024-01-11")

len(dicts)
```

```python
df = {
    "time":[],
    "sigma_w":[],
    "sigma_p":[],
    "w_rhat":[]
}

for d in dicts:
    #print("==========")
    #print(f"n = {d['n']}, m = {d['m']}, ratio = {np.log2(d['m'])/d['n']}")
    #print(d["accuracy_distribution"])
    #print(d.keys())
    df["time"].append(d["time"])
    df["sigma_w"].append(d["sigma_w"])
    df["sigma_p"].append(d["sigma_p"])
    df["w_rhat"].append(np.mean(d["w_rhat"]))

df = pd.DataFrame(df).round(3)
df["σw"] = df["sigma_w"]
df["σu"] = df["sigma_p"]
df = df.groupby(["σw", "σu"]).mean().reset_index()
df = df.pivot(index="σw", columns="σu", values="w_rhat")
```

```python
sns.heatmap(df)
plt.savefig("Convergence_hyperparameters.png")
```

```python
file_1 = 'Eunomia/experiments/SX2/2023-11-19/n-10_k-2_m-256_sigma_w-0.1_sigma_p-1_n_samples-1500.yaml'
file_2 = "Eunomia/experiments/SX2/2023-11-19/n-10_k-2_m-64_sigma_w-0.1_sigma_p-1_n_samples-1500.yaml"
file_3 = "Eunomia/experiments/SX2/2023-11-19/n-10_k-2_m-128_sigma_w-0.1_sigma_p-1_n_samples-1500.yaml"



with open(file_1, 'r') as file_1:
    data_1 = yaml.safe_load(file_1)
    
    
with open(file_2, 'r') as file_2:
    data_2 = yaml.safe_load(file_2)
    
    
with open(file_3, 'r') as file_3:
    data_3 = yaml.safe_load(file_3)
data_1.keys()
```

```python
data = {
    "accuracy":[],
    "m":[],
}
for d1,d2,d3 in zip(data_1["accuracy_distribution"], data_2["accuracy_distribution"], data_3["accuracy_distribution"]):
    data["accuracy"].append(d1)
    data["m"].append(256)
    
    data["accuracy"].append(d2)
    data["m"].append(64)
    
    data["accuracy"].append(d3)
    data["m"].append(128)
    
    
data["ratio"] = np.log(data["m"]) / 10
data = pd.DataFrame(data)
data["ratio"] = data["ratio"].round(2)
```

```python
sns.histplot(data=data, x="accuracy", kde=True, hue = "ratio", palette = "Spectral")
plt.savefig("accuracy_hist.PNG")
```

```python
file_1 = 'Eunomia/experiments/SX2/2023-11-19/n-10_k-2_m-8_sigma_w-0.1_sigma_p-1_n_samples-1500.yaml'
file_2 = "Eunomia/experiments/SX2/2023-11-19/n-10_k-2_m-64_sigma_w-0.1_sigma_p-1_n_samples-1500.yaml"
file_3 = "Eunomia/experiments/SX2/2023-11-19/n-10_k-2_m-128_sigma_w-0.1_sigma_p-1_n_samples-1500.yaml"



with open(file_1, 'r') as file_1:
    data_1 = yaml.safe_load(file_1)
    
    
with open(file_2, 'r') as file_2:
    data_2 = yaml.safe_load(file_2)
    
    
with open(file_3, 'r') as file_3:
    data_3 = yaml.safe_load(file_3)
```

```python
data = {
    "accuracy":[],
    "m":[],
}

for d1,d2,d3 in zip(data_1["kt_d"], data_2["kt_d"], data_3["kt_d"]):
    data["accuracy"].append(d1)
    data["m"].append(8)
    
    data["accuracy"].append(d2)
    data["m"].append(64)
    
    data["accuracy"].append(d3)
    data["m"].append(128)
    
    
data["ratio"] = data["m"] / (np.ones_like(data["m"])*(2**10))
data = pd.DataFrame(data)
data["Percentage of the dataset"] = data["ratio"].round(2)

data["Kendall-Tau Distance"] = data["accuracy"]
```

```python
sns.histplot(data=data, x="Kendall-Tau Distance", kde=True, hue = "% total coalitions", palette = "Spectral")
plt.savefig("kt_hist.PNG")
```
