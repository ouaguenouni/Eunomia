```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ROADEF_TESTS/test_n3_p0.825.csv")
np.unique(df.n.values)
```

```python
df_4 = df[df.n == 4].groupby(["d_g", "d_m"]).mean().reset_index()[["d_g", "d_m", "%r"]].astype(float)
df_4["Model's interactions"] = df_4["d_m"]
df_4["Graph's interactions"] = df_4["d_g"]
df_4 = df_4.pivot(index="Model's interactions", columns="Graph's interactions", values="%r")
df_4.sort_index(level=0, ascending=False, inplace=True)
sns.heatmap(df_4, annot=True, fmt=".3f")
plt.savefig("interactions_4.PNG")
```

```python
df_4 = df[df.n == 5].groupby(["d_g", "d_m"]).mean().reset_index()[["d_g", "d_m", "%r"]].astype(float)
df_4["Model's interactions"] = df_4["d_m"]
df_4["Graph's interactions"] = df_4["d_g"]
df_4 = df_4.pivot(index="Model's interactions", columns="Graph's interactions", values="%r")
df_4.sort_index(level=0, ascending=False, inplace=True)
sns.heatmap(df_4, annot=True, fmt=".3f")
plt.savefig("interactions_5.PNG")
```

```python
import os
ch = "ROADEF_TESTS"
dfs = []
for ch in sorted(os.listdir(ch)):
    n = (ch.split("_")[1][1:])
    p = ch.split("_")[2][1:-4]
    df = pd.read_csv(os.path.join("ROADEF_TESTS", ch))
    df["n"] = int(n)
    df["p"] = float(p)
    dfs.append(df)
```

```python
df = pd.concat(dfs).reset_index()
sns.set_style ("whitegrid")
```

```python
df_g = df.groupby(["d_g", "d_m", "%r"]).mean().reset_index()[["d_g", "d_m", "%r"]]
df_g = df_g.pivot_table(index = "d_m", columns = "d_g", values = "%r")
df_g.sort_index(level=0, ascending=False, inplace=True)
sns.heatmap(df_g, annot=True, fmt=".3f")
```

```python
import os
ch = "ROADEF_TESTS_2"
dfs = []
for ch in sorted(os.listdir(ch)):
    n = (ch.split("_")[1][1:])
    p = ch.split("_")[2][1:-4]
    df = pd.read_csv(os.path.join("ROADEF_TESTS_2", ch))
    df["n"] = int(n)
    df["p"] = float(p)
    dfs.append(df)
```

```python
df = pd.concat(dfs).reset_index()
sns.set_style ("whitegrid")
df = df[df.n == 5]
df = df[df.d_m <= 5]
df = df[df.d_g <= 5]
```

```python
df_g = df[df.n == 5].groupby(["d_g", "d_m", "%r"]).mean().reset_index()[["d_g", "d_m", "%r"]]
df_g = df_g.pivot_table(index = "d_m", columns = "d_g", values = "%r")
df_g.sort_index(level=0, ascending=False, inplace=True)
sns.heatmap(df_g, annot=True, fmt=".3f")
```
